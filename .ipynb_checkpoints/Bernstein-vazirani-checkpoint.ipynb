```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secret_number= '111101'
```

```python
circuit=QuantumCircuit(len(secret_number)+1,len(secret_number))
circuit.h(range(len(secret_number)))
circuit.x(len(secret_number))
circuit.h(len(secret_number))
circuit.barrier()

j=0
for i in list(reversed(secret_number)):
    print (i)
    if i=='1':
        circuit.cx(j,len(secret_number))
    j+=1
        
circuit.barrier()

circuit.h(range(len(secret_number)))
circuit.barrier()

circuit.measure(range(len(secret_number)),range(len(secret_number)))
circuit.draw(output='mpl')
```

```python
simulator= Aer.get_backend('qasm_simulator')
results=execute(circuit, backend=simulator,shots=1).result()
print (results.get_counts())
plot_histogram(results.get_counts())
```
